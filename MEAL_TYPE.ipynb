```python
Ingredients = []

for i in range(len(df1['Ingredients'])):
    ing = re.sub(",","",df['Ingredients'][i]).strip()
    ing = ing.replace("[","").replace("]","").replace("'","").replace("(","").replace(")","").replace("  "," ")
    ing = re.sub("[0-9]","",ing)
    #ing = lambda: map(str.strip, ing)
    Ingredients.append(ing)
    
df1['Ingredients'] = Ingredients
```

```python
Food = []

for i in range(len(df1['Food_Name'])):
    ing = re.sub(",","",df1['Food_Name'][i]).strip()
    ing=ing.lower()
    #ing = lambda: map(str.strip, ing)
    Food.append(ing) 
    
df1['Food_Name'] = Food
```

```python
NV = ['chicken','seafood','beef','barbecue','brisket','turkey','salami','ham','pork','rib',
      'poultry','fish','shrimp','lobster','crab','turkey','octopus','scampi','flounder','prawns','egg','eggs']

#res = any(nv in df['Ingredients'][0] for nv in NV)
def meal_type(string,lst):

    res = any(nv in string for nv in NV)
    if res == False:
        return lst.append('Vegetarian')
    else:
        return lst.append('Non-Vegetarian')
    
Meal=[]

for i in range(len(df1['Ingredients'])):
    meal_type(df1['Ingredients'][i],Meal)  
```
