```python
urls = food2['links']
print(urls[0])
```

```python
from recipe_scrapers import scrape_me

title = []
ingredients = []
img = []
nutrients = []

for u in range(len(food2['links'])):
    
    scraper = scrape_me(urls[u])
    print (u)
        
    try:
        title.append(scraper.title())
    except:
        title.append('n/a')
        
    try:
        ingredients.append(scraper.ingredients())
    except:
        ingredients.append('n/a')
        
    try:
        img.append(scraper.image())
    except:
        img.append('n/a')
        
    try:
        nutrients.append(scraper.nutrients())
    except:
         nutrients.append('n/a')    
```

```python
#For 1st 3257 entries
Mix=pd.DataFrame({'Food_Name' : title, 'Ingredients': ingredients, 'Image':img, 'Nutrients':nutrients, 'Entree':'Main', 'Cuisine':'Mexican', 'Course':'Lunch Dinner','Meal_Type':'Non-Vegetarian'})
#Mix
cal = []
Carb = []
Protien =[]
fats=[]

for i in range(len(Mix['Nutrients'])):
    res = Mix['Nutrients'][i]
    try:
        cal.append(res['calories'])
    except:
        cal.append('NA')
    
    try:
        Carb.append(res['carbohydrateContent'])
        
    except:
        Carb.append('NA')
    try:
        Protien.append(res['proteinContent'])
    except:
        Protien.append('NA')
    try:
        fats.append(res['fatContent'])
    except:
        fats.append('NA')
```
